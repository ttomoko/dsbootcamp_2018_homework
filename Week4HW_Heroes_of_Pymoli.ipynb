```python
#dependencies
import pandas as pd
import numpy as np
```

```python
#file to load
file_to_load = "purchase_data.csv"
purchase_data= pd.read_csv(file_to_load,error_bad_lines=False)
#purchase_data2=purchase_data.copy()
purchase_data.head()
```

```python
purchase_data.shape
```

```python
#player count

total_player = len(purchase_data["SN"].unique())
total_player_df=pd.Series({
    "Total Players": total_player,
})
total_player_df
```

```python
#Purchasing Analysis (total)

#Number of Unique Items
unique = len(purchase_data["Item ID"].unique())

#Average Purchase Price
average_purchase_price = purchase_data["Price"].mean()
average_purchase_price

#Total Number of Purchases
total_number_purchases = purchase_data["Purchase ID"].count()
total_number_purchases


#Total Revenue
total_revenue= purchase_data["Price"].sum()
total_revenue

purchasing_analysis_df= pd.DataFrame({
   "Number of Unique Items": unique,
    "Average Price": "$"+ str(average_purchase_price),
    "Number of Purchases": [total_number_purchases],
    "Total Revenue": "$"+ str(total_revenue)
})
purchasing_analysis_df
```

```python
#Gender Demographics
gender_demo=purchase_data.drop_duplicates(["SN"])["Gender"].value_counts()
avg_players=gender_demo*100/total_player
sum_gender=pd.DataFrame({
    "Total Count": gender_demo,
    "Percentage of Players": avg_players})
sum_gender
```

```python
#Purchasing Analysis (Gender)
purchase_count=pd.DataFrame(purchase_data["Gender"].value_counts())

total_purchase_gender=purchase_data[["Gender", "Price"]]
total_purchase=total_purchase_gender.groupby(["Gender"]).sum()

avg_purchase_price=total_purchase["Price"]/purchase_count["Gender"]
#avg_purchase_price.columns=["Average Purchase Price"]
avg_ttl_purchase=total_purchase["Price"]/sum_gender["Total Count"]


sum_gender2=pd.DataFrame({
   "Purchase Count": purchase_count['Gender'],
   "Total purchase": total_purchase['Price'],
   "Average Purchase Price": avg_purchase_price,
   "Average TTL Price": avg_ttl_purchase
})
sum_gender2
```

```python
#Age Demographics
purchase_data.drop_duplicates(["SN"], inplace = True)
purchase_data["Categories"] = pd.DataFrame(pd.cut(purchase_data["Age"], [-np.inf,10,15,20,25,30,35,40,+np.inf],labels=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+'],right = False, include_lowest = True))
agecount = pd.DataFrame(purchase_data.groupby("Categories").count(), index =['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+'])

summary_age = pd.DataFrame()
summary_age["Total Count"] = agecount["Purchase ID"]
summary_age["Percentage of Players"] = (summary_age["Total Count"] / summary_age["Total Count"].sum())*100
#summary_age2 = summary_age.copy()
#summaryAgeCpy.index.name = 'Tabulador'
summary_age
```
